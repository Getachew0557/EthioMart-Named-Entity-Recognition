#### remove telegram link and different emojo

```python
import pandas as pd
import re

# Function to remove emojis, Telegram links, the word "·â¥·àå·åç·à´·àù", and the first \n
def clean_text(text):
    # Remove emojis using regex
    text = re.sub(r'[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF\U0001F680-\U0001F6FF\U0001F700-\U0001F77F\U0001F900-\U0001F9FF\U0001FA00-\U0001FA6F\U00002702-\U000027B0]+', '', text)
    # Remove Telegram links
    text = re.sub(r't\.me/\S+', '', text)
    # Remove the word "·â¥·àå·åç·à´·àù"
    text = re.sub(r'·â¥·àå·åç·à´·àù', '', text)
    # Remove leading newline characters and other unnecessary spaces
    text = text.lstrip()
    return text

# Apply the cleaning function to the 'Message' column
df['Message'] = df['Message'].apply(clean_text)

# Display the cleaned DataFrame
print(df['Message'])
```

#### Remove different tags and contents on Reactions entity

```python
import pandas as pd
import re

# Function to extract the number of reactions
def extract_reaction_count(reaction_str):
    # Check if the input is a string
    if isinstance(reaction_str, str):
        # Use regex to find all numbers in the string
        numbers = re.findall(r'\d+', reaction_str)
        # Return the count of reactions
        return len(numbers)
    else:
        # Return 0 if the input is not a string or is NaN
        return 0

# Apply the extraction function to the 'Reactions' column
df['Reactions'] = df['Reactions'].apply(extract_reaction_count)

# Display the updated DataFrame with the number of reactions
print(df['Reactions'])
```
